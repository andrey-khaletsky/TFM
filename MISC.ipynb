```python
import os
```

```python
filenames = [filename for filename in os.listdir('gtfs-book') if '.md' in  filename and '~' not in filename]
filenames
```

```python
filenames = [(int(filename.split('-', maxsplit=2)[1]), 
           filename.split('-', maxsplit=2)[2]) for filename in filenames]
filenames
```

```python
from  operator import itemgetter
filenames.sort(key=itemgetter(0))
```

```python
filenames
```

```python
filenames = [f'ch-{filename[0]:02d}-{filename[1]}' for filename in filenames]
filenames
```

```python
output = [open(f'gtfs-book/{filename}', 'r').read() for filename in filenames]
output
```

```python
output = '\n\n\n'.join(output)
output
```

```python
with open('gtfs-book.md', 'w') as file:
    file.write(output)
```
